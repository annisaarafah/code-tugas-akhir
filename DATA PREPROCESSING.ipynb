```python
import numpy as np
import pandas as pd

import dill
import pickle

import  matplotlib as mpl
import  matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import rcParams
```

```python
data = pd.read_csv("GSE9006_data_set.csv")
```

```python
data
```

```python
data.columns
```

```python

data.drop(columns=['Type'], inplace = True)
```

```python
data.groupby('label')['label'].count()
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
selector = VarianceThreshold()
selector.fit_transform(data)
```

```python
sel_feature = selector.get_feature_names_out()
```

```python
data = data[sel_feature]
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(data, test_size=0.2,  stratify= data['label'])
```

```python
test
```

```python
train.to_csv('data_train.csv', index=None)
```
